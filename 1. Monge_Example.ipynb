# 1. Transport map

**Example 1.2**

```python
# import modules
import numpy as np
import matplotlib.pyplot as plt
import ot
import ot.plot
from ot.datasets import make_1D_gauss as gauss
```

```python
# Generate the data

n = 100  # nb bins

# bin positions
x = np.arange(n, dtype=np.float64)

# Gaussian distributions
source_dist = gauss(n, m=20, s=5)  # m = mean, s = std
target_dist = gauss(n, m=60, s=10)

# cost matrix
cost_matrix = ot.dist(x.reshape((n, 1)), x.reshape((n, 1)))
cost_matrix /= cost_matrix.max()

# Compute transportation matrix
transport_map = ot.emd(source_dist, target_dist, cost_matrix)
```

```python
# Plot the source distribution
plt.figure(1,figsize=(10,6))
plt.plot(x, source_dist, color='b', label = 'Source distribution')
plt.fill_between(x, source_dist, color = 'b', alpha=0.3)

# Plot the target distribution
plt.plot(x, target_dist, color='r', label = 'Target distribution')
plt.fill_between(x, target_dist, color='r', alpha=0.3)

# plt.plot(x,np.zeros(len(x)),color='black')

# Plot the transportation map as a segment
for i in range(n-1):
    plt.plot([x[i], x[transport_map[i].argmax()]], [source_dist[i], target_dist[transport_map[i].argmax()]], 'k-', alpha=0.5)
plt.plot([x[n-1], x[transport_map[n-1].argmax()]], [source_dist[n-1], target_dist[transport_map[n-1].argmax()]], 'k-', alpha=0.5, label='Transportation map')

plt.legend()
plt.show()
```
